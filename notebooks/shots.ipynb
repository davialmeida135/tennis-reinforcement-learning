```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the parsed CSV file
parsed_csv_path = "../data/processed/parsed_charting-m-points-2010s.csv"
df = pd.read_csv(parsed_csv_path)

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())

print("\nBasic Statistics:")
print(df.describe())

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values[missing_values > 0])

# Plot the distribution of shot types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="shot_type", order=df["shot_type"].value_counts().index)
plt.title("Distribution of Shot Types")
plt.xlabel("Shot Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the distribution of shot types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="last_shot_type", order=df["last_shot_type"].value_counts().index)
plt.title("Distribution of Last Shot Types")
plt.xlabel("Last Shot Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the distribution of shot directions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="shot_direction", order=df["shot_direction"].value_counts().index)
plt.title("Distribution of Shot Directions")
plt.xlabel("Shot Direction")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analyze the relationship between shot type and whether it is a winner
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="shot_type", hue="is_winner", order=df["shot_type"].value_counts().index)
plt.title("Shot Type vs. Winner")
plt.xlabel("Shot Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Is Winner")
plt.tight_layout()
plt.show()

# Analyze the relationship between shot type and whether it is an error
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="shot_type", hue="is_error", order=df["shot_type"].value_counts().index)
plt.title("Shot Type vs. Error")
plt.xlabel("Shot Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Is Error")
plt.tight_layout()
plt.show()
```
